```python
# Create a simple File
```

```python
with open("sample.txt","w") as f:
    f.write("Name: Omkar\n")
    f.write("Role: Data Scientist\n")
    f.write("Location: India\n")
```

```python
# reading the file 
```

```python
with open("sample.txt","r") as f:
    data = f.read()
    print(data)
```

```python
# reading line by line
```

```python
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
    
```
