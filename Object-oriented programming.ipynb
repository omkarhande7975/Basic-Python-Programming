```python
# Static Methods
```

```python
class Calculator:
    @staticmethod
    def mul(a,b):
        return a*b
```

```python
Calculator.mul(10,20)
```

```python
# Encapsulation
```

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
```

```python
BankAccount(1000).get_balance()
```

```python
# single Inheritance
```

```python
class Animal:
    def eat(self):
        print("Eating")
class Dog(Animal):
    pass
```

```python
Dog().eat()
```

```python
class Studentmarks:
    def marks(self):
        print("Marks")
class Studentpercent(Studentmarks):
    pass
```

```python
# Method Overriding
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")
class Car(Vehicle):
    def start(self):
        print("Car is starting...")
```

```python
car=Car()
```

```python
car.start()
```
