```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4]
y= [10,20,25,30]
```

```python
# Basic line plot
```

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)

plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
```

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x= [1,2,3]

y1=[2,4,6]
y2=[1,3,5]

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
categories = ["A","B","C"]
values = [10,20,15]
plt.bar(categories, values)
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Simple Bar Chart")
plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

```python
labels = ["A","B","c"]
sizes = [30,40,30]
```

```python

labels = ["A", "B", "C"]
sizes = [30, 40, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Simple Pie Chart")
plt.show()
```

```python


scores = [42, 45, 48, 52, 55, 58, 62, 65, 70, 75]

counts, bins, patches = plt.hist(scores, bins=5, edgecolor='black', color='skyblue')

plt.xticks(bins)

plt.bar_label(patches)

plt.title("Histogram of Student Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
x= [1,2,3,4,5,6,7]
y= [2,4,5,4,5,7,8]
```

```python
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot Example")
plt.show()
```
