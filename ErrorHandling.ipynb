```python
# Try-except
```

```python
# Try contains code that might cause an error. if an error happens, except handels it.
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    y = 60 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# multiple except blocks
# Different error can be handled in diferent ways using miltiple except blocks.
```

```python
try:
    num = int("abc")
    result = 10 / num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("Division by zero error")
```

```python
try:
    num = 10 / 0
    result = 10 / num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("Division by zero error")
```

```python
# 3. Multiple exception in one except
```

```python
# You can handle more than one error type in a single block.
```

```python
try:
    x = int("abc")
except (ValueError, TypeError):
    print("Value or Type Error occurred")
```

```python
# 4. Generic Exception
# catches any kind of Error when You don't specify the except type. 
```

```python
try:
    x = int("abc")
except Exception as e:
    print("Error occurred:",e)
```

```python
try:
    x = 10 / 0
except Exception as e:
    print("Error occurred:",e)
```

```python
# 5. Else block
# Runs only when no error occurs in the try block.
```

```python
try:
    x = 10 / 2
except ZeroDivsionError:
    print("Error")
else:
    print("Success", x)
```

```python
# 6. finally Block
# Run every time, whether an Error happens or not. Often used for cleanup.
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Error occured")
finally:
    print("This always run")
```
