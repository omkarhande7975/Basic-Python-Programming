```python
# Positional Arguments
```

```python
def show(name, age):
    print(name, age)
show("omkar",19)    
```

```python
def name_age(name,age):
    return f"My name is {name}, I am {age} years old."
name_age("omkar",19)    
```

```python
name_age(age = 19, name = "omkar")
```

```python
def name_age_marks(name,age,marks):
    print(f"my name is {age} and i am {name} years old.i get {marks} in java.")
name_age_marks("omkar",19,80)    
```

```python
# **kwargs (key value)
```

```python
def info(**data):
    print(data)

info(name="Rahul", age=19)    
```

```python
# *args (Multiple values)
```

```python
def add(*nums):
    print(sum(nums))

add(10, 20, 30)   
```

```python
def multiplication (*arg):
    multi = 1
    for number in arg:
        multi *= number
    return multi

multiplication(10,20,30)
```

```python
# lambda Function 
```

```python
add = lambda a, b: a + b
print(add(10, 20))
```

```python
#  check wheather number is even or odd
```

```python
numbers = [1,2,3,4,5,6,8]

count = 0

for num in numbers:
    if num % 2 == 0:
        print(num, "is Even")
        count +=1
    else:
        print(num, "is Odd")

print("Total Even Numbers:", count)        
```
